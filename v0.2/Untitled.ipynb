```python
# load all the mu=90 cells
ap4 = metaclass.twiss('./4_dipole/twiss0_ap.out')
ap3 = metaclass.twiss('./3_dipole/twiss0_ap.out')
ap2 = metaclass.twiss('./2_dipole/twiss0_ap.out')
ap1 = metaclass.twiss('./1_dipole/twiss0_ap.out')
tw4 = metaclass.twiss('./4_dipole/twiss0.out')
tw3 = metaclass.twiss('./3_dipole/twiss0.out')
tw2 = metaclass.twiss('./2_dipole/twiss0.out')
tw1 = metaclass.twiss('./1_dipole/twiss0.out')
apmin90 = np.array([np.min(ap1.N1),np.min(ap2.N1),np.min(ap3.N1),np.min(ap4.N1)])
betmin90 = np.array([np.min(tw1.BETX),np.min(tw2.BETX),np.min(tw3.BETX),np.min(tw4.BETX)])
```

```python
# load all the mu=60 cells
ap4_60 = metaclass.twiss('./4_dipole/twiss1_ap.out')
ap3_60 = metaclass.twiss('./3_dipole/twiss1_ap.out')
ap2_60 = metaclass.twiss('./2_dipole/twiss1_ap.out')
ap1_60 = metaclass.twiss('./1_dipole/twiss1_ap.out')
tw4_60 = metaclass.twiss('./4_dipole/twiss1.out')
tw3_60 = metaclass.twiss('./3_dipole/twiss1.out')
tw2_60 = metaclass.twiss('./2_dipole/twiss1.out')
tw1_60 = metaclass.twiss('./1_dipole/twiss1.out')
apmin60 = np.array([np.min(ap1_60.N1),np.min(ap2_60.N1),np.min(ap3_60.N1),np.min(ap4_60.N1)])
betmin60 = np.array([np.min(tw1_60.BETX),np.min(tw2_60.BETX),np.min(tw3_60.BETX),np.min(tw4_60.BETX)])
```

```python
# aperture vs number of dipoles per cell
dipNum = np.arange(1,5,1)
fig, ax = plt.subplots(1)
ax.plot(dipNum, apmin90,'.',label='mu=90')

ax.plot(dipNum, apmin60,'.',label='mu=60')

ax.legend()
```

```python
# beta vs number of dipoles per cell
dipNum = np.arange(1,5,1)
fig, ax = plt.subplots(1)
ax.plot(dipNum, betmin90,'.',label='mu=90')

ax.plot(dipNum, betmin60,'.',label='mu=60')

ax.legend()
```
